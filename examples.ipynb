```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
```

```python
#do a X-gate on zero qubit
#1- defind the quantum circuite where is 1 regester
qc=QuantumCircuit(1)
#applied X-gate with zero qubit
qc.x(0)
qc.draw('mpl')
```

```python
#the resulte
backend = Aer.get_backend('statevector_simulator')
out = execute(qc,backend).result().get_statevector()
print(out)
```

```python
#y-gate z-gate
qc.y(0)
qc.z(0)
qc.draw('mpl')
```

```python
# creat circuit
qc= QuantumCircuit(3)
#aplied Hadamard Gate on the circuit
for qubit in range (3):
    qc.h(qubit)
#see teh circuit
qc.draw('mpl')
```

```python
#identity gate
qc.i(0)
qc.draw('mpl')
```
